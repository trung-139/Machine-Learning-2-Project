```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
df
```

```python
y = df["Churn"]
enc = LabelEncoder()
y = enc.fit_transform(y)

data = df.iloc[:, 1:-1]
df["Churn"]=enc.fit_transform(df["Churn"])
```

```python
c = (data.dtypes == "object")
catcol = list(c[c].index)
c
```

```python
for col in catcol:
  data[col] = enc.fit_transform(data[col])
```

```python
from scipy import stats
```

```python
#define feature
features=["gender","SeniorCitizen","Partner","Dependents","tenure","PhoneService","MultipleLines","InternetService","OnlineSecurity","OnlineBackup","DeviceProtection","TechSupport","StreamingTV","StreamingMovies","Contract","PaperlessBilling","PaymentMethod","MonthlyCharges","TotalCharges"]
target = 'Churn'
important_features=[]
# calculate the t-test and p-value
# perform a hypothesis test for each feature
for feature in features:
    # separate the data into groups based on the target variable
    group1 = data[df["Churn"] == 1][feature]
    group2 = data[df["Churn"] == 0][feature]

    # calculate the t-test and p-value
    t, p = stats.ttest_ind(group1, group2)

    # print the results
    print(feature)
    print("t-value: ", t)
    print("p-value: ", p)
    print("")
    if p < 0.05:
        important_features.append(feature)
        print(feature, "is an important predictor with p-value:", p)
    else:
        print(feature, "is not an important predictor with p-value:", p)
# t-test is a statistical test used to compare the means of two independent groups. It is used to determine if there is a significant difference between the means of the two groups being compared.
#Specifically, the ttest_ind() function used in the code performs an independent samples t-test, which assumes that the two groups being compared are independent of each other. The function takes in two sets of data as input and returns both the t-value and the associated p-value.

# The t-value is a statistic calculated from the data that measures the difference between the means of the two groups in terms of standard error units. The p-value is the probability of obtaining a t-value as extreme or more extreme than the one observed, assuming that there is no difference between the means of the two groups.

# By comparing the calculated p-value to a predetermined level of significance (e.g., 0.05), we can determine whether the observed difference between the two groups is statistically significant or not. If the p-value is less than the level of significance, we reject the null hypothesis and conclude that there is a significant difference between the means of the two groups. If the p-value is greater than the level of significance, we fail to reject the null hypothesis and conclude that there is no significant difference between the means of the two groups.




```
